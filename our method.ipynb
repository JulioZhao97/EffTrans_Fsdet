<font size=10>calculate statistics of base class</font>

```python
import os
import pickle
import numpy as np

base_class_path = '/data/zhaozhiyuan/mmfewshot-main/base_statistics/2022-02-28/11:33:07'
classes = os.listdir(base_class_path)
print(classes)
base_statistics = {}
for c in classes:
    feats = os.listdir(os.path.join(base_class_path, c))
    embeds = []
    # all features
    for feat in feats:
        f = open(os.path.join(base_class_path, c, feat), 'rb')
        embed = pickle.load(f)
        embeds.append(embed)
        f.close()
    
    # mean & variance
    dim = 1024
    mean = np.zeros(embeds[0].shape)
    for embed in embeds:
        mean += embed
    mean /= len(embeds)
    
    variance = np.zeros((dim,dim))
    for embed in embeds:
        minus = np.expand_dims((embed - mean), axis = 1)
        v = np.matmul(minus, minus.T)
        variance += v
    variance /= (len(embeds) - 1)
    base_statistics[c] = {}
    base_statistics[c]['mean'] = mean
    base_statistics[c]['variance'] = variance
    
f = open(os.path.join(base_class_path, 'base_statistics.pt'), 'wb')
pickle.dump(base_statistics, f)
f.close()
```
